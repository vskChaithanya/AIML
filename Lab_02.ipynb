```python
import numpy as np
from queue import Queue
```

```python
graph = {
    0:[1,2,3],
    1:[0,2],
    2:[0,1,3,5],
    3:[0,2,4],
    4:[3,5,7],
    5:[4,6,7],
    6:[5,7],
    7:[4,5,6],
    8:[]
}
print("The adjacency List representing the Graph is  : ")
print(graph)
```

```python
#Define the BFS function
def bfs(graph,source):
  Q=Queue()
  visited_vertices=set()
  Q.put(source)
  visited_vertices.add(source)
  while not Q.empty():
    vert=Q.get()
    print(vert,end="-->> ")
    for u in graph[vert]:
      if u not in visited_vertices:
        Q.put(u)
        visited_vertices.add(u)

print("the bfs traversal of the graph with source 0 : ")
bfs(graph,0)
```
